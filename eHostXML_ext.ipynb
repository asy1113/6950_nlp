```python
# extract social isolation notes

def ehostxml(filename='filename', class_checker = {'class_checker'}):
    '''Extract social isolation notes from eHost xml'''
    tree = ET.parse(filename)
    root = tree.getroot()
    file_name = root.attrib['textSource']
    #print(file_name)
    
    return_check='no_note'
    
    txt=set()
    list1=[]
    for j in root.iter('mentionClass'):
        class_id = j.attrib['id']
        class_id1 = set(class_id.lower().split('-'))
        #-----
        list2=[] # remove possibe spaces
        for j1 in class_id1:
            st = j1.strip()
            list2.append(st)
        class_id1=set(list2)
        #-----
        #print(class_id)
        
        if class_checker.issubset(class_id1):
            txt.add(file_name)
            txt.add(j.text)
            class_id2 = class_id
            #print(file_name)
            #print ('---', j.text)
            
    if len(txt)>1:
        list1=list(txt)
        list1.sort()
        print(class_id2)
        print(list1)   
        print('-----')
        return_check='yes_note'
        
    return return_check, file_name
```

```python
try:
    import xml.etree.cElementTree as ET
except ImportError:
    import xml.etree.ElementTree as ET
import csv
import chardet
import os
path = "C:/Users/HuaiZhong/2018/ann_xml"
files = os.listdir(path)
#filename="123_456-789.txt.knowtator.xml"
class_checker = {'social isolation'}
print('Number of notes')
print(len(files))
#note1=[]

count = 0
pos_files=[]
neg_files=[]
pos_txt_files=[]
neg_txt_files=[]
for file in files[:]:
    filename=os.path.join(path,file)
    iso, txt_file=ehostxml(filename=filename, class_checker = class_checker)
    if iso == 'yes_note':
        count=count+1
        pos_files.append(filename)
        pos_txt_files.append(txt_file)
    else:
        neg_files.append(filename)
        neg_txt_files.append(txt_file)
print('Number of notes with social isolation --- ', str(count))
```

```python
# copy file
import shutil
def copyFile(src, dest):
    try:
        shutil.copy2(src, dest)        
    except shutil.Error as e:
        print('Error: %s' % e)
    except IOError as e:
        print('Error: %s' % e.strerror)
```

csv to target_rules

```python
import csv
row_all=[]
with open('KB/socialIsolation_terms_simple_040418.csv') as csvfile:
    csv_file = csv.reader(csvfile, delimiter=',')
    for row in csv_file:
        #print(row[0]+ " --- " +row[1])
        row_all.append(row)
print(len(row_all))

rules1=[]
for row in row_all:
    rules1.append("Comments: ''")
    rules1.append("Direction: ''")
    rules1.append("Lex: "+ row[1])
    rules1.append("Regex: "+ row[0])
    rules1.append("Type: ISOLATION")
    rules1.append("---")

print(len(rules1))
rules='\n'.join(rules1)
print(rules)
        
with open('KB/social_target_rules.yml', 'w') as f:
    f.write(rules)
```
